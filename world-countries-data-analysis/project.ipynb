```python
country_by_region = df.Region.value_counts()
country_by_region
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']= (12,12)
matplotlib.rcParams['figure.facecolor']= '#00000000'
```

```python
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.barplot(country_by_region.index,country_by_region)
ax.set_xlabel('Region')
ax.set_ylabel('Count')
plt.title('Number of countries in each region')
plt.xticks(rotation=85)
for i in ax.patches:
    width = i.get_width()
    height = i.get_height()
    ax.text(x= i.get_x() + width/2,
            y= height,
            s='{:.0f}'.format(height),
            ha='center')
```

## Q2> Describe the area of all the countries for each using some Geo-graph .

```python
country_df = pd.read_csv('./proj_data/country.csv')
just_name = country_df[['CountryCode','Country']]
type(just_name)
just_name
```

```python
#I do this because original data has some spaces in values of country column, so by strip method I cut it out those extra space thus why i can manipulate with no errors
df['Country']=df.Country.str.strip()
df.Country.unique()
```

```python
new_df = df.merge(just_name, on='Country')
new_df.head()
```

```python
import folium
country_geo = './proj_data/world-countries.json'
country_geo
```

```python
plot_data_to = new_df[['CountryCode','Country',df.columns[3]]]
```

```python
m = folium.Map([43, -100], zoom_start=0)

choropleth = folium.Choropleth(
    geo_data=country_geo,
    data=plot_data_to,
    columns=["CountryCode", plot_data_to.columns[2]],
    key_on="feature.id",
    fill_color="YlGnBu", fill_opacity=0.7, line_opacity=0.2,
    legend_name='Area square mile'
).add_to(m)

# The underlying GeoJson and StepColormap objects are reachable
print(type(choropleth.geojson))
print(type(choropleth.color_scale))

folium.LayerControl(collapsed=False).add_to(m)

m


###Note that
#this <folium map graph>only work with chrome browser , if u open it with edge u will only see blank . Open in chrome and  see how it works.
```

## Q3> what is the literacy rate by region and average literacy in globe?

```python
ax = sns.distplot(df['Literacy (%)'],kde=False,color='r')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    ax.text(x= p.get_x() + width/2,
                   y = height,
                   s='{:.0f}'.format(height),
                   ha='center')
plt.axvline(df['Literacy (%)'].mean(),color='k')
plt.text(x=df['Literacy (%)'].mean()+0.5,
            y=max([i.get_height() for i in ax.patches]),
            s='mean: {:.0f}'.format(df['Literacy (%)'].mean()),
            ha = 'center'
        )
```

## Q4> Compare the birhrate and deathrate of the countries of ASIA region.

Here more than hundred countries have the literacy rate above 90%.
Globally, average literacy rate is 83%.


```python
asia = df[df.Region.str.contains('ASIA')]
asia_df = asia[['Country','Deathrate','Birthrate']]
list_country = list(asia_df.Country)
asia_df
```

```python
index = np.arange(len(list_country))
bar_width = 0.5
fig, ax = plt.subplots()

birthrate = ax.bar(index, asia_df['Birthrate'], bar_width, label='Brithrate')
Deathrate = ax.bar(index+bar_width, asia_df['Deathrate'], bar_width, label='Deathrate')

ax.set_title('Brithrate vs Deathrate for each country')
ax.set_xticklabels(list_country)
ax.set_xticks(index+ bar_width/2)
ax.legend()
plt.xticks(rotation=75)


for p in ax.patches:
    width= p.get_width()
    height = p.get_height()
    ax.text(x=p.get_x() + (width/2)+0.1,
               y = height,
               s= '{:.0f}'.format(height),
               ha= 'center',
               fontsize=10,
               rotation=0,
               color= 'purple')
    
plt.show()
```
